```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta, date
import os
import math
import networkx as nx
import matplotlib as mlt
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap as Basemap
from matplotlib.patches import Polygon
from matplotlib.collections import PatchCollection
from matplotlib.patches import PathPatch
import matplotlib.patches as patches
import shapefile
import plotly.express as px
import plotly.graph_objects as go
from matplotlib.colors import ListedColormap, LinearSegmentedColormap
from cmcrameri import cm
import seaborn as sns
```

```python
Df_hc=pd.read_csv('Cluster_HydrologicalDrought.csv')
```

```python
att_ca=pd.read_csv(f'All_attributes_drought.csv')
Df_att_1=pd.merge(att_ca, Df_hc, on='id', how='inner')
Df_att_1=Df_att_1.rename(columns={"catch_area":"catchment area"})
```

```python
for i in range(512):
    cluster1 = Df_att_1.loc[lambda Df_att_cluster: Df_att_cluster['cluster'] == 0]
    cluster2 = Df_att_1.loc[lambda Df_att_cluster: Df_att_cluster['cluster'] == 1]
    cluster3 = Df_att_1.loc[lambda Df_att_cluster: Df_att_cluster['cluster'] == 2]
    cluster4 = Df_att_1.loc[lambda Df_att_cluster: Df_att_cluster['cluster'] == 3]
    cluster5 = Df_att_1.loc[lambda Df_att_cluster: Df_att_cluster['cluster'] == 4]
```

### Drougth characteristics

```python
##Boxplot Connectedness
plt.figure(figsize=(6,4))
Number_events=pd.concat([cluster1['connectedness'].reset_index(drop=True),
                         cluster2['connectedness'].reset_index(drop=True),
                         cluster3['connectedness'].reset_index(drop=True),
                         cluster4['connectedness'].reset_index(drop=True),
                         cluster5['connectedness'].reset_index(drop=True),
                        ], ignore_index=True,axis=1)

Number_events.columns=["C1","C2","C3","C4","C5"]
Number_events.head()

cor={"C1":"#fecc5c",
"C2":"#f03b20",
"C3":"#bd0026",
"C4":"#fd8d3c",
"C5":'#ffffb2'}

sns.set_style("whitegrid",{"grid.color": ".8", "grid.linestyle": ":"})

NE=sns.boxplot(data=Number_events, palette=cor, width=0.6, linewidth=1.5, fliersize=3, showfliers = False)
NE.set_ylabel("Connectedness", fontsize = 18)
NE.tick_params(labelsize=14)
NE.figure.savefig("Connectedness.png", dpi=300)
plt.show()
```
